```python
print("Soal 1")

# Baca data dari file excel
df_excel = pd.read_excel("dataset/jml_produksi_sampah.xlsx", sheet_name="data")
df_excel['total_produksi_tahunan'] = df_excel.apply(lambda baris: baris['jumlah_produksi_sampah'] * 365, axis=1)

# Group data berdasarkan tahun dan kota, dan hitung total produksi tahunan
df_grouped = df_excel.groupby(['nama_kabupaten_kota', 'tahun'])['total_produksi_tahunan'].sum().unstack()

# Menampilkan data yang telah digroup
print(df_grouped)

# Plot diagram batang vertikal
df_grouped.plot(kind='bar', figsize=(18, 6), color=['skyblue', 'salmon', 'green', 'orange', 'purple'])  # Sesuaikan warna untuk setiap kota
plt.ylabel('Jumlah Produksi (Ton)')
plt.xlabel('Nama Kota/Kabupaten')
plt.title('Jumlah Produksi Tahunan per Kota')
plt.gca().yaxis.set_major_formatter(tck.StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.show()
```
